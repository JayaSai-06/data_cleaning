```python
import pandas as pd
import numpy as np
import re

df = pd.read_csv("data.csv")
df.columns = df.columns.str.replace(r'\xa0', ' ', regex=True).str.strip()
money_columns = ['Actual gross', 'Adjusted gross (in 2022 dollars)', 'Average gross']

def clean_money(val):
    if pd.isna(val):
        return np.nan
    val = re.sub(r'\[.*?\]', '', str(val))
    val = re.sub(r'[^\d.]', '', val)
    return pd.to_numeric(val, errors='coerce')

for col in money_columns:
    df[col] = df[col].apply(clean_money)
peak_col=['Peak','All Time Peak']
for col in peak_col:
    df[col] = df[col].str.replace(r'\D', '', regex=True)
    df[col] = pd.to_numeric(df[col], errors='coerce')

df['Tour title'] = (
    df['Tour title']
    .astype(str)
    .str.replace(r'\[.*?\]', '', regex=True)
    .str.replace(r'[†‡]', '', regex=True)
    .str.strip()
)
if 'Ref.' in df.columns:
    df.drop(columns=['Ref.'], inplace=True)

print("Missing values summary:\n", df.isnull().sum())
df['Peak'] = df['Peak'].fillna(df['Peak'].median())
df['All Time Peak'] = df['All Time Peak'].fillna(df['All Time Peak'].median())
print("Missing values summary:\n", df.isnull().sum())
df.head()
```
